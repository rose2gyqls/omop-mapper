## Join

```python
import pandas as pd

# 1. 데이터 로드 (타입 안정성 확보)
concept = pd.read_csv('/Users/rose/omop-mapper/data/omop-cdm/CONCEPT.csv', sep='\t', low_memory=False)
syn = pd.read_csv('/Users/rose/omop-mapper/data/omop-cdm/CONCEPT_SYNONYM.csv', sep='\t', low_memory=False)
syn = syn[syn['language_concept_id'].isin([4180186])]

# ID 타입을 정수로 통일
concept['concept_id'] = concept['concept_id'].astype(int)
syn['concept_id'] = syn['concept_id'].astype(int)

# 2. 원본 데이터 구분자 추가
concept['name_type'] = 'Original'

# 3. 룩업 테이블 생성 (에러 수정 지점)
# concept_id, concept_name, name_type을 '제외'한 나머지 컬럼만 추출합니다.
target_cols = [c for c in concept.columns if c not in ['concept_id', 'concept_name', 'name_type']]

# concept_id를 인덱스로 잡고, 나머지 메타데이터 컬럼들만 보존합니다.
concept_lookup = concept.drop_duplicates('concept_id').set_index('concept_id')[target_cols]

# 4. 동의어 행 생성 및 메타데이터 결합
syn_rows = syn[['concept_id', 'concept_synonym_name']].copy()
syn_rows.rename(columns={'concept_synonym_name': 'concept_name'}, inplace=True)
syn_rows['name_type'] = 'Synonym'

# 5. join을 통해 메타데이터 병합
# syn_rows의 concept_id와 concept_lookup의 index(concept_id)를 매칭합니다.
syn_rows = syn_rows.join(concept_lookup, on='concept_id')

# 6. 최종 합치기 및 확인
final_df = pd.concat([concept, syn_rows], ignore_index=True)

# 잘 들어갔는지 특정 ID로 테스트 (예: 45756805)
print(final_df[final_df['concept_id'] == 45756805])
```
